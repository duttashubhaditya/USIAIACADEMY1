# Content based Recommendation

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3, max_features=None, 
                      strip_accents='unicode', analyzer='word', token_pattern = r'\w{1,}',
                     ngram_range = (1,3), stop_words = 'english')

mov_df2['overview'] = mov_df2['overview'].fillna('')
```

```python
tvf_matrix = tfv.fit_transform(mov_df2['overview'])
```

```python
print(tvf_matrix)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(tvf_matrix,tvf_matrix)
```

```python
sig
```

```python
indices = pd.Series(mov_df2.index, index = mov_df2['original_title']).drop_duplicates()
indices
```

```python
def recommend(title, sig = sig ):
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores, key = lambda x: x[1], reverse = True)
    sig_scores = sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    return mov_df2['original_title'].iloc[movie_indices]
```

```python
recommend('Spy Kids')
```

# Nearest neighbor item by collaborative filtering

```python
mov_ndf = pd.read_csv('movies.csv', usecols = ['movieId','title'], dtype = {'movieId':'int32','title':'str'})
rat_ndf = pd.read_csv('ratings.csv', usecols = ['userId','movieId','rating'],
                      dtype = {'userId':'int32','movieId':'int32','rating':'float32'})

mov_ndf,rat_ndf                     
```

```python
mov_masterdf = rat_ndf.merge(mov_ndf,on = 'movieId')
mov_masterdf.head()
```

```python
mov_rating = mov_masterdf.dropna(axis = 0, subset = ['title'])
mov_rating_count = (mov_rating.groupby(by = 'title')['rating'].count().reset_index().rename(columns = {'rating':
                                                                                                       'ratings_count'})[['title',
                                                                                                                          'ratings_count']])
mov_rating_count
```

```python
mov_rating_comb_count = mov_masterdf.merge(mov_rating_count, left_on='title', right_on='title')
mov_rating_comb_count
```

```python
mov_rating_count['ratings_count'].describe()
```

```python
rated_popular_mov = mov_rating_comb_count.query('ratings_count>=50')
rated_popular_mov
```

```python
mov_features_df = rated_popular_mov.pivot_table(index='title', columns='userId', values='rating').fillna(0)
mov_features_df
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
mov_features_df_mat = csr_matrix(mov_features_df.values)

model_knn = NearestNeighbors(metric= 'cosine', algorithm='brute')
model_knn.fit(mov_features_df_mat)
```

```python
query_index = np.random.choice(mov_features_df.shape[0])
X = mov_features_df.iloc[query_index,:].values.reshape(1,-1)
distances, indices = model_knn.kneighbors(X, n_neighbors=6)
```

```python
for i in range (0, len(distances.flatten())):
    if i == 0:
        print ('Recommendation for {0}:\n'.format(mov_features_df.index[query_index]))
    
    else:
        print('{0}: {1}, with distance of {2}:'.format(i,mov_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
